```python
import re
import nltk
from nltk.stem import PorterStemmer
import pandas as pd
import numpy as np
```

```python
# removing puntuations and generating stop word count.

from collections import Counter
import operator
tokens=[]
punctuations='''~()-[]{}“”";:\\|/,.<>@$''^+*%&=?▪✅『』🌡【★✔】�►✔️"®™’——✿⌚⚡👉※●🎧🏆💡☀🏋️⭐🌕☀⛄🔥🎁☑️ž¤#▶🧡'''
for i in range(1,1315):
    if i==797:
        continue
    with open(r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/Amazon Scraped Documents/document"+str(i), 'r', encoding="utf8") as fp:
        # read line number 3 to 5
        # index starts from 0
        document_lines=fp.readlines()[0].lower()
        nopunct_line=""
        for char in document_lines:
            if char not in punctuations:
                nopunct_line+=char
        tokens.extend(nopunct_line.split(" "))

c=Counter(tokens)
c=sorted(c.items(), key=lambda item: item[1],reverse=True)
print(c)
```

```python
# Generating top 10 stopwords list
stop_words=[]
count=10
for i in c:
    if not count:
        break
    stop_words.append(i[0])
    count-=1  
print(stop_words)
```

```python
# removing stopwords from the text files and storing them in another folder.
corpus=[]
for i in range(1,1315):
    if i==797:
        continue
    with open(r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/Amazon Scraped Documents/document"+str(i), 'r', encoding="utf8") as fp:
        # read line number 3 to 5
        # index starts from 0
        document_lines=fp.readlines()[0].lower()
        corpus.append(document_lines)
    nopunct_line=""
    for j in document_lines.split(" "):
        for char in j:
            if char not in punctuations:
                nopunct_line+=char
        nopunct_line+=" "
    final_line=""
    for j in nopunct_line.split(" "):
        if j not in stop_words:
            final_line+=j
        final_line+=" "
    file=open(r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/stop_words_removed/document"+str(i),"w", encoding="utf-8")
    file.write(final_line)
    print(str(i), final_line,"\n")
    file.close()
    
```

```python
######################################

def stemming(content):
    
    stem = PorterStemmer()

    content = re.sub('[^A-Za-z]+', ' ', content)
    tokens = nltk.word_tokenize(content)
  
    stem_tokens = [stem.stem(word) for word in tokens]
    clean_stem_tokens = ' '.join(map(str,  stem_tokens))

    shortword = re.compile(r'\W*\b\w{1,2}\b')
    clean_stem_tokens = shortword.sub('', clean_stem_tokens)

    
    return clean_stem_tokens
```

```python
#########################################
for i in range(1,1315):
    if i==797:
        continue
    with open(r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/stop_words_removed/document"+str(i), 'r', encoding="utf-8") as f:
        content = f.readlines()
        content=" ".join(content)
        new_content = stemming(content)
        with open(r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/stemmed_docs/document"+str(i), 'w', encoding="utf-8") as f1:
            f1.write(new_content)
            f1.close()
    f.close()
```

```python
def cosine_similarity(input_user):
    punctuations_remove_user_input=""
    for i in input_user.split(" "):
        for char in i:
            if char not in punctuations:
                punctuations_remove_user_input+=char
        punctuations_remove_user_input+=" "
    punctuations_remove_user_input=punctuations_remove_user_input.lower()
    #print(punctuations_remove_user_input)
    stopword_remove_user_input=""
    for i in punctuations_remove_user_input.split(" "):
        if i not in stop_words:
            stopword_remove_user_input+=i
        stopword_remove_user_input+=" "
    #print(stopword_remove_user_input)
    word_tokens = nltk.word_tokenize(stopword_remove_user_input)
    fdict_user_input = dict(nltk.FreqDist(word_tokens))
    #print(fdict_user_input)
    q_df=pd.DataFrame([fdict_user_input])
    q_df = q_df.replace(np.nan, 0)
    u=q_df.transpose()
    u.reset_index(inplace=True)
    cosinesim=[]
    for j in range(len(vocab_df)):
        y=vocab_df.iloc[j:j+1].transpose()
        y.reset_index(inplace=True)
        df3=pd.merge(y,u, how='outer',on="index")
        df3.columns=["index","Df","Qf"]
        df3 = df3.replace(np.nan, 0)
        cos_sim = np.dot(df3["Df"],df3["Qf"])/(np.linalg.norm(df3["Df"])*np.linalg.norm(df3["Qf"]))
        if round(cos_sim,4) not in [x[1] for x in cosinesim]:
            cosinesim.append((j+1,round(cos_sim,4)))
    cosinesim.sort(key=lambda a: a[1],reverse=True)
    #print(cosinesim[:10])
    output_cos=""
    for i in cosinesim[:10]:
        j=i[0]
        if i[0]>796:
            j+=1
        with open (r"C:/Users/Dell/Information Retrival Project (Amazon search engine)/Amazon Scraped Documents/document"+str(j)) as f:
            output_cos+="Printing document "+str(j)+"\n"
            for line in f:
                output_cos+=line+"\n"
            output_cos+="\n"
    return output_cos
```

```python
import tkinter as tk

# Top level window
frame = tk.Tk()
frame.title("TextBox Input")
frame.geometry('400x200')
#frame.resizable(False, False)
# Function for getting Input
# from textbox and printing it
# at label widget

def printInput():
    inp = inputtxt.get(1.0, "end-1c")
    lbl.insert(tk.END,f"Provided Input: "+cosine_similarity(inp))
    
l1=tk.Label(frame,text="Enter your query :", borderwidth=10,font=12)
l1.pack()
# TextBox Creation
inputtxt = tk.Text(frame,height = 5,width = 20)

inputtxt.pack()

# Button Creation
printButton = tk.Button(frame,text = "Print",command = printInput)
printButton.pack()

# Label Creation
lbl = tk.Text(frame, height = 500,width = 700)
lbl.pack()

# create a scrollbar widget and set its command to the text widget
scrollbar = tk.Scrollbar(frame, orient='vertical', command=lbl.yview)
#scrollbar.grid(row=0, column=1, sticky=tk.NS)

#  communicate back to the scrollbar
lbl['yscrollcommand'] = scrollbar.set

frame.mainloop()
```
