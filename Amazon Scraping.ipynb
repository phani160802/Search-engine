```python
#### This program scrapes naukri.com's page and gives our result as a
#### list of all the job_profiles which are currently present there.

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from collections import deque
from urllib.parse import urlparse

def getnextpage(soup):
    # this will return the next page URL
    if not soup:
        return
    pages = soup.find('div', {"class":"a-section a-text-center s-pagination-container"})
    url = 'https://www.amazon.in' + str(pages.find('span', {'class':"s-pagination-strip"}).find_all('a')[-1]['href'])
    return url


# initiating the webdriver. Parameter includes the path of the webdriver.
browser = webdriver.Chrome('./chromedriver')

# load the webpage
browser.get('https://www.amazon.in')

browser.maximize_window()

# get the input elements
input_search = browser.find_element('id','twotabsearchtextbox')
search_button = browser.find_element('xpath',"(//input[@type='submit'])[1]")

# send the input to the webpage
input_search.send_keys("Smartphones under 40000")
search_button.click()
count=0
# this is just to ensure that the page is loaded

frontier_queue=deque()
url_set=set()
title_set=set()

while(True):
    html = browser.page_source
    
    # Now, we could simply apply bs4 to html variable
    soup = BeautifulSoup(html, "html.parser")
    all_divs = soup.find_all('a', {"class":"a-link-normal s-no-outline"})
    
    for link in all_divs:
        browser.get('https://www.amazon.in'+link.get('href'))
        
        if browser.title in title_set:
            continue
        title_set.add(browser.title)           
        browser.back()
        url_set.add('https://www.amazon.in'+link.get('href'))
        frontier_queue.append('https://www.amazon.in'+link.get('href'))
    browser.get(getnextpage(soup))
    count+=1
    if count==15:
        break


browser.close() # closing the webdriver
print(frontier_queue)
```

### Fetching suggested urls from the link by popping the url from queue

```python
# initiating the webdriver. Parameter includes the path of the webdriver.
browser = webdriver.Chrome('./chromedriver')
while(frontier_queue):
    # load the webpage
    browser.get(frontier_queue.popleft())
    
    browser.maximize_window()
    
    html = browser.page_source
    
    # Now, we could simply apply bs4 to html variable
    soup = BeautifulSoup(html, "html.parser")
    all_divs = soup.find_all("div",{"class":"a-section sp_offerVertical p13n-asin sp_ltr_offer"})
    # print(all_divs)
    for link in all_divs:
        if not link.find('a',{"class":"a-link-normal adReviewLink a-text-normal"}):
            continue
        
        if len(url_set)==1200:
            break
        suggestedurl="https://www.amazon.in"+link.find('a',{"class":"a-link-normal adReviewLink a-text-normal"})["href"]
        page_title=link.find('div',{"class":'sponsored-products-truncator-truncated'}).getText()
        if page_title not in title_set:
            url_set.add(suggestedurl)
            title_set.add(page_title)
            frontier_queue.append(suggestedurl)
    
browser.close() # closing the webdriver
```

## Writing Product description into file

```python
url_list=list(url_set)
```

```python
# initiating the webdriver. Parameter includes the path of the webdriver.
browser = webdriver.Chrome('./chromedriver')
i=0
for j in (url_set):
    
    browser.get(j)
    
    browser.maximize_window()
    
    html = browser.page_source
    
    # Now, we could simply apply bs4 to html variable
    soup = BeautifulSoup(html, "html.parser")
    all_divs = soup.find("ul",{"class":"a-unordered-list a-vertical a-spacing-mini"})    
    if not all_divs:
        continue
    i+=1
    file=open("Amazon Documents/document"+str(i),"w", encoding="utf-8")
    file.write(browser.title)
    file.write(all_divs.get_text()+"\n")
    file.write(j)
    file.close()
file.close()
browser.close() # closing the webdriver
```
